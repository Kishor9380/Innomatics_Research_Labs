Smart Light Controller

```python

class SmartLight:

    # Initialize light with name and default status OFF
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    # Turn the light ON
    def turn_on(self):
        self.status = "ON"

    # Turn the light OFF
    def turn_off(self):
        self.status = "OFF"

    # Display current status
    def display_status(self):
        print(self.name + " is " + self.status)


light_name = input("Enter Light Name: ")
light = SmartLight(light_name)

action = input("Enter Action (ON/OFF): ")

if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()
else:
    print("Invalid Action")

light.display_status()
```

Employee ID Card System

```python
class Employee:

    # Initialize employee details
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # Display employee ID card details
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
department = input("Enter Department: ")

employee = Employee(name, emp_id, department)
employee.display_id_card()
```

Mobile Contact Record

```python
class Contact:

    # Initialize contact details
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    # Display contact information
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


name = input("Enter Contact Name: ")
phone = input("Enter Phone Number: ")

contact = Contact(name, phone)
contact.display_contact()
```

Product Price Tag Generator

```python
class Product:

    # Initialize product details
    def __init__(self, name, price):
        self.name = name
        self.price = price

    # Display formatted price tag
    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


name = input("Enter Product Name: ")
price = input("Enter Price: ")

product = Product(name, price)
product.display_price_tag()
```

Movie Rating Display System

```python
class Movie:

    # Initialize movie details
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    # Display movie details
    def display_details(self):
        print("Movie:", self.name)
        print("Rating:", str(self.rating) + " / 5")


name = input("Enter Movie Name: ")
rating = input("Enter Rating: ")

movie = Movie(name, rating)
movie.display_details()
```

Delivery Address Manager

```python
class Delivery:

    # Initialize delivery details
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    # Display delivery details
    def display_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


name = input("Enter Customer Name: ")
address = input("Enter Address: ")

delivery = Delivery(name, address)
delivery.display_details()
```
